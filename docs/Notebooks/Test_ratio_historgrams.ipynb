# Test ratio historgrams

This notebook is to test how a linear relationship of a particular parameter will affect the historgram distribution of the ratio between itself and the center value. 

```python
import matplotlib.pyplot as plt
import numpy as np
def linear(x, m, b):
    "Return line."
    return m*x + b
```

```python
values = np.array([-1, -0.5, 0, 0.5, 1])
npoints = 5000
sigma = 0.05
```

```python
m = 0.5
m1=m
b = 1
x = np.linspace(-1,1, 20)

plt.figure(figsize=(10,7))
plt.subplot(211)
plt.plot(x, linear(x, m, b), "--")
plt.plot(values, linear(values, m, b), "o")
plt.xlabel("{Fe/H]")
plt.title("y = {0} [Fe/H] + {1}".format(m, b))

# mean y value at each fe/h value
mu = linear(values, m, b)
# random sample 
feh1 = sigma * np.random.randn(npoints, len(mu)) + mu
plt.plot(values, feh1.T ,".")
ratios = (feh1[:,[0,1,3,4]] / feh1[:,2][:,np.newaxis]).T

plt.subplot(212)
plt.hist(ratios[0], bins=20, alpha=0.5)
plt.hist(ratios[1], bins=20, alpha=0.5)
plt.hist(ratios[2], bins=20, alpha=0.5)
plt.hist(ratios[3], bins=20, alpha=0.5)
plt.legend(labels=["{0} / 0.0".format(val) for val in values[[0,1,3,4]]], title="[Fe/H]")
plt.xlim((0,2.5))
```

Randomly sampling 500 points with normali distribution at each metalicity.
These are to roughly represent the broadening and spectral type combinations for the rv presions.

The 4 historgrams are offset depending on the mean value of the linear function at each metallicity. and are relative to the middle value at [Fe/H] = 0.0
